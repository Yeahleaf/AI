```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X = np.random.rand(100, 2)
```

```python
X[0]
```

```python
plt.scatter(X[:,0], X[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
clf = KMeans(n_clusters=3)
```

```python
clf.fit(X)
```

```python
len(clf.labels_)
```

```python
plt.scatter(X[:,0], X[:,1], c=clf.labels_)
```
