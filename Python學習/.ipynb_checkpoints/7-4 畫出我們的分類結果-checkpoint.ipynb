```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.svm import SVC
```

```python
xx = np.linspace(-3, 5, 30)
yy = np.linspace(-8, 6, 30)
```

```python
X, Y = np.meshgrid(xx, yy)
```

```python
x = X.ravel()
y = Y.ravel()
```

```python
plt.scatter(x, y)
```

```python
bb = [1, 2, 3, 4]
cc = [5, 6, 7, 8]
```

```python
list(zip(bb,cc))
```

```python
clf = SVC()
```

```python
A = [[1, 8], [6, -7], [3, 3], [-6, -7]]
B = [1, 1, 2, 2]
```

```python
clf.fit(A,B)
```

```python
clf.predict([[-6, -7]])
```

```python
C = list(zip(x, y))
```

```python
D = clf.predict(C)
```

```python
plt.scatter(x, y, s=50, c=D)
```
