```python
## 第一題_算閏年：

def year(num):
    if (num % 4 == 0) and ((num % 100 != 0) or (num % 400 == 0)):
        print("西元" + str(num) + "年是閏年")
    else:
        print("西元" + str(num) + "年不是閏年")
```

```python
year(1000)
```

```python
year(1200)
```

# 挑戰題一_Bubble Sort

```python
grades = np.random.randn(100)
grades = 10*grades + 70
grades
```

```python
for i in range(len(grades)-1):
    for j in range(len(grades)-1-i):
        if grades[j] > grades[j+1]:
            mark = grades[j]
            grades[j] = grades[j+1]
            grades[j+1] = mark
```

```python
grades
```
