```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston.feature_names
```

```python
X = boston.data
Y = boston.target
```

```python
plt.figure(figsize=(10, 10))
for i, feature in enumerate(boston.feature_names):
    plt.subplot(5, 3, i+1)
    plt.scatter(X[:,i], Y, s = 1)
    plt.xlabel(feature)
    plt.ylabel("Price")
    plt.tight_layout()
```
